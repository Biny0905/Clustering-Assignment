##### Hopkins Statistics:

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan

def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

### K-means:

##### Silhouette Analysis:

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(df_km)
    sse_.append([k, silhouette_score(df_km, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

##### Sum of Sqr Distances:

```python
ss = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(df_km)
    ss.append(model_clus.inertia_)

plt.plot(ss)
```

We will take K = 4, as we can see the highest peak is at 4 (k=4) and in sum of sqr distances graph.

##### Kmeans with K as 4

#### Hierachical Clustering:

```python

# heirarchical clustering single method
df_clustering = df_final
fig = plt.figure(figsize = (12,12))
mergings = linkage(df_clustering, method = "single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

We can see that single method heirachical clustering things are not that clear , thus we will do complete heirachical clustering method

```python

# HC with complete method
fig = plt.figure(figsize = (11,11))
mergings = linkage(df_clustering, method = "complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

Cutting Dendogram vertically such that n_clusters = 4

```python
clusterCut = pd.Series(cut_tree(mergings, n_clusters = 4).reshape(-1,))
```

```python
country_hc = pd.concat([country_df.reset_index(), clusterCut], axis=1)
country_hc.columns = ['country', 'child_mort','exports','health','imports','income','inflation','life_expec','total_fer','gdpp',
                  'ClusterID']
country_hc.head()
```
